Добавьте новый столбец под названием «время до следующей станции».<br>
    Чтобы получить это значение, мы вычитаем время станций для пар смежных станций.<br>
    Мы можем вычислить это значение без использования оконной функции SQL, но это может быть очень сложно.<br>
    Проще это сделать с помощью оконной функции LEAD .<br>
    Эта функция сравнивает значения из одной строки со следующей строкой, чтобы получить результат.<br> В этом случае функция сравнивает значения в столбце «время» для станции со станцией сразу после нее.
